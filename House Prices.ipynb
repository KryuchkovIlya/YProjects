**Lets replace missings in `living_area` with median_value of area by number of rooms**

```python
df.value_counts('rooms')
```

```python
def column_median(cat, data):
    for index in df[cat].unique():
        area_median = df.loc[df[cat] == index, data].median()
        df.loc[df[cat] == index, data] = df.loc[df[cat] == index, data].fillna(area_median)   
column_median(cat='rooms', data='living_area')
```

```python
df['living_area'].isna().sum()  # Now there's no missing values
```

**I'll create a function which creates a new column where all the localities are written with a capitall letter**

```python
def place_category(place_series):
    for title in place_series:
         if str.istitle(title):
                return place_series[place_series.find(title):]
```

```python
df['locality_name_short'] = df['locality_name'].apply(place_category) # apply the function
```

```python
df['locality_name_short'].unique()
```

**I'll add a new column with price of one square meter**

```python
df['sqr_price'] = df.last_price / df.total_area
```

```python
df['weekday_name'] = df['first_day_exposition'].dt.dayofweek
df['month_name'] = df['first_day_exposition'].dt.month
df['year_name'] = df['first_day_exposition'].dt.year
```

**Function with a floor type**

```python
def floor_category(floor, floors_total):
    if floor == floors_total:
        return 'последний'
    elif floor == 1 or floor == 0:
        return 'первый'
    else:
        return 'другой'
```

```python
df['floor_category'] = df.apply(lambda x: floor_category(x['floor'], x['floors_total']),
                        axis=1)
```

```python
df['city_centers_nearest_km'] = round(df['cityCenters_nearest'] / 1000) # the distance to center
```
